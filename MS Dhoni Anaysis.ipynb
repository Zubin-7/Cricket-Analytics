```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the CSV file into a Pandas DataFrame
df = pd.read_csv('data.csv', usecols=[
    'season', 'start_date', 'striker', 'runs_off_bat', 'ball', 'wicket_type'
])
```

```python
# Extract year from start_date
df['year'] = df['start_date'].str[:4].astype(int)
```

```python
# Filter data for MS Dhoni's innings
dhoni_innings = df[df['striker'] == 'MS Dhoni']
```

```python
# Group by year and calculate cumulative runs, balls, and dismissals for MS Dhoni
dhoni_performance_over_years = dhoni_innings.groupby('year').agg({
    'runs_off_bat': 'sum',
    'ball': 'sum',
    'wicket_type': 'count'
})
```

```python
# Calculate average and strike rate for MS Dhoni
dhoni_performance_over_years['average'] = dhoni_performance_over_years['runs_off_bat'] / dhoni_performance_over_years['wicket_type']
dhoni_performance_over_years['strike_rate'] = (dhoni_performance_over_years['runs_off_bat'] / dhoni_performance_over_years['ball']) * 100
```

```python
# Filter data for the last few years (e.g., last 3 years)
last_few_years_data = df[df['year'] >= (df['year'].max() - 2)]
```

```python
# Group by batsman and calculate cumulative runs, balls, and dismissals for all batsmen in 2021
all_batsmen_2021 = last_few_years_data[last_few_years_data['year'] == 2021].groupby('striker').agg({
    'runs_off_bat': 'sum',
    'ball': 'sum',
    'wicket_type': 'count'
})
```

```python
# Calculate average and strike rate for all batsmen in 2021
all_batsmen_2021['average'] = all_batsmen_2021['runs_off_bat'] / all_batsmen_2021['wicket_type']
all_batsmen_2021['strike_rate'] = (all_batsmen_2021['runs_off_bat'] / all_batsmen_2021['ball']) * 100
```

```python
# Display the analysis for MS Dhoni over the years
print("MS Dhoni's Performance Analysis Over the Years:")
print(dhoni_performance_over_years)
```

```python
# Display the analysis for all batsmen in 2022
print("\nAll Batsmen Performance Analysis in 2022:")
print(all_batsmen_2021)
```

```python
# Plotting the analysis for MS Dhoni over the years
plt.figure(figsize=(12, 6))
plt.plot(dhoni_performance_over_years.index, dhoni_performance_over_years['runs_off_bat'], marker='o', label='Runs')
plt.title("MS Dhoni's Runs Over the Years")
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(dhoni_performance_over_years.index, dhoni_performance_over_years['average'], marker='o', label='Average')
plt.title("MS Dhoni's Batting Average Over the Years")
plt.xlabel('Year')
plt.ylabel('Average')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(dhoni_performance_over_years.index, dhoni_performance_over_years['strike_rate'], marker='o', label='Strike Rate')
plt.title("MS Dhoni's Strike Rate Over the Years")
plt.xlabel('Year')
plt.ylabel('Strike Rate')
plt.legend()
plt.show()
```

```python
# Extract year from start_date
df['year'] = df['start_date'].str[:4].astype(int)
```

```python
# Filter data for the year 2022
year_2021_data = df[df['year'] == 2021]
year_2021_data
```

```python
# Group by batsman and calculate cumulative runs, balls, and dismissals for the top 15 batsmen
top_15_batsmen = year_2021_data.groupby('striker').agg({
    'runs_off_bat': 'sum',
    'ball': 'sum',
    'wicket_type': 'count'
}).sort_values(by='runs_off_bat', ascending=False).head(15)

top_15_batsmen
```

```python
# Calculate average and strike rate for the top 15 batsmen
top_15_batsmen['average'] = top_15_batsmen['runs_off_bat'] / top_15_batsmen['wicket_type']
top_15_batsmen['strike_rate'] = (top_15_batsmen['runs_off_bat'] / top_15_batsmen['ball']) * 100
top_15_batsmen['average']
```

```python
# Filter data for MS Dhoni's innings in the year 2022
dhoni_innings_2021 = year_2021_data[year_2021_data['striker'] == 'MS Dhoni']
```

```python
# Calculate cumulative runs, balls, and dismissals for MS Dhoni
dhoni_2021_performance = dhoni_innings_2021.agg({
    'runs_off_bat': 'sum',
    'ball': 'sum',
    'wicket_type': 'count'
})
```

```python
# Combine the performance data for the top 15 batsmen and MS Dhoni in 2022
all_batsmen_2021_performance = pd.concat([top_15_batsmen, pd.DataFrame([dhoni_2021_performance], index=['MS Dhoni'])])
```

```python
# Plotting the analysis for all batsmen and MS Dhoni in 2022
plt.figure(figsize=(12, 8))

# Adjust marker size based on runs for top 15 batsmen
marker_size_batsmen = all_batsmen_2021['runs_off_bat'] * 0.01

# Adjust marker size based on runs for MS Dhoni
marker_size_dhoni = dhoni_2021_performance['runs_off_bat'] * 0.01

# Use a fixed color for all markers (top 15 batsmen)
plt.scatter(
    all_batsmen_2021['average'],
    all_batsmen_2021['strike_rate'],
    s=marker_size_batsmen,
    color='blue',  # You can choose any color
    alpha=0.7,
    label='Top 15 Batsmen'
)

# Use a different color for MS Dhoni's marker
plt.scatter(
    dhoni_2021_performance['average'],
    dhoni_2021_performance['strike_rate'],
    s=marker_size_dhoni,
    color='red',  # You can choose any color
    alpha=0.7,
    label='MS Dhoni'
)

plt.title('Batsmen Performance in 2021')
plt.xlabel('Average')
plt.ylabel('Strike Rate')

# Display the name of each batsman next to their respective point for top 15 batsmen
for i, txt in enumerate(all_batsmen_2021.index):
    plt.annotate(txt, (all_batsmen_2021['average'][i], all_batsmen_2021['strike_rate'][i]), fontsize=8)

# Display the name of MS Dhoni's point
plt.annotate('MS Dhoni', (dhoni_2021_performance['average'], dhoni_2021_performance['strike_rate']), fontsize=8)

# Add legend
plt.legend()

plt.show()
```
